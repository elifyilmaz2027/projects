```
'''
cont. 1) Find Taylor series expansions for tan(x) and arctan(x)
'''

import sympy as sy
import numpy as np
from sympy.functions import tan,atan
import matplotlib.pyplot as plt

plt.style.use("ggplot")

x = sy.Symbol('x')


# Factorial function
def factorial(n):
    if n <= 0:
        return 1
    else:
        return n*factorial(n-1)

# Taylor approximation at x0 of the function 'f'
def taylor(f,x0,n):
    i = 0
    p = 0
    while i <= n:
        p = p + (f.diff(x,i).subs(x,x0))/(factorial(i))*(x-x0)**i
        i += 1
    return p
print("Taylor series expansion for f(x)=tanx")
f = tan(x)
print(taylor(f,0,5))

print("Taylor series expansion for f(x)=arctanx")
f = atan(x)
print(taylor(f,0,5))
```

```
'''10) Trapezoid Rule Trapezoid Uniform( function, a, b, n) exercise 5.2.1'''

import numpy as np
import matplotlib.pyplot as plt


def trapezoidal_uniform_function(function, a, b, n):
    x = np.linspace(a, b, n + 1)
    y = function(x)
    y_right = y[1:]
    y_left = y[:-1]
    dx = (b - a) / n
    T = (dx / 2) * np.sum(y_right + y_left)
    return T


# FUNCTIONS FROM 5.2.2
def function_1(x):
    return np.sin(x)


def function_2(x):
    return np.exp(x)


def function_3(x):
    return np.arctan(x)


print("trapezoid section\n")
print("Trapezoid Rule for sinx")
print(trapezoidal_uniform_function(function_1, 0, np.pi, 50))
print("\nTrapezoid Rule for exp(x)")
print(trapezoidal_uniform_function(function_2, 0, 1, 50))
print("\nTrapezoid Rule for arctan(x)\n")
print(trapezoidal_uniform_function(function_3, 0, 1, 50))


# Romberg Integration

def romberg_integrator(function, a, b, n):
    romberg_array = np.array([[0] * (n + 1)] * (n + 1), float)
    h_step = b - a
    romberg_array[0, 0] = 0.5 * h_step * (function(a) + function(b))

    power_2 = 1
    for i in range(1, n + 1):
        h_step = 0.5 * h_step

        sum = 0.0
        power_2 = 2 * power_2
        for k in range(1, power_2, 2):
            sum = sum + function(a + k * h_step)

        romberg_array[i, 0] = 0.5 * romberg_array[i - 1, 0] + sum * h_step

        power_4 = 1
        for j in range(1, i + 1):
            power_4 = 4 * power_4
            romberg_array[i, j] = romberg_array[i, j - 1] + (romberg_array[i, j - 1] - romberg_array[i - 1, j - 1]) / (power_4 - 1)

    return romberg_array


# FUNCTIONS FROM 5.3.2
def function_4(x):
    return 1 / (1 + x)


def function_5(x):
    return np.exp(x)


def function_6(x):
    return 1 / (1 + x ** 2)


print("ROMBERG - APPLICATION")
print("\nromberg for 1/(1+x)")
print(romberg_integrator(function_4, 0, 1, 5))
print("\nromberg for exp(x)")
print(romberg_integrator(function_5, 0, 1, 5))
print("\nromberg for (1+x**2)")
print(romberg_integrator(function_6, 0, 1, 5))


def simpsons_function(function, a, b, number):
    if number % 2 == 1:
        raise ValueError("Not a suitable number. It must be divisible to 2.")
    dx = (b - a) / number
    x = np.linspace(a, b, number + 1)
    y = function(x)
    S = dx / 3 * np.sum(y[0:-1:2] + 4 * y[1::2] + y[2::2])
    return S


print("Simpsons Section\n")

print(simpsons_function(function_1, 0, np.pi, 50))
print(simpsons_function(function_2, 0, 1, 50))
print(simpsons_function(function_3, 0, 1, 50))


# ADAPTIVE SIMPSON
def adaptive_simpson_function(function, a, b, epsilon, level, level_max):
    h = (b - a)
    x0 = a
    x1 = a + 0.25 * h
    x2 = a + 0.5 * h
    x3 = a + 0.75 * h
    x4 = b

    f0 = function(x0)
    f1 = function(x1)
    f2 = function(x2)
    f3 = function(x3)
    f4 = function(x4)

    one_step_simps = h * (f0 + 4.0 * f2 + f4) / 6.0
    two_steps_simps = h * (f0 + 4.0 * f1 + 2.0 * f2 + 4.0 * f3 + f4) / 12.0

    if (level >= level_max):
        result = two_steps_simps
        print("maximum level to be attained")


    else:
        if (abs(two_steps_simps - one_step_simps) <= epsilon * 15):
            result = two_steps_simps + (two_steps_simps - one_step_simps) / 15

        else:
            result = adaptive_simpson_function(function, x0, x2, epsilon / 2, level + 1, level_max) + adaptive_simpson_function(function, x2, x4, epsilon / 2, level + 1,
                                                                                                                                level_max)

    return result


# FUNCTIONS FROM 6.1.1
def function_7(x):
    return 4 / (1 + x ** 2)


def function_8(x):
    return 8 * (np.sqrt(1 - x ** 2) - x)


eps = 0.5 * (10 ** (-5))
level_max = 4
lev = 1

print("\nadaptive simpsons 4/(1+x**2)\n")
print(adaptive_simpson_function(function_7, 0, 1, eps, lev, level_max))

print("\nadaptive simpsons 8*(np.sqrt(1-x**2)-x)\n")
print(adaptive_simpson_function(function_8, 0, 1 / np.sqrt(2), eps, lev, level_max))

'Plot 3.1.13 Graph for f7 and f8'

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
inx = np.arange(-10, 10, 0.1)
f7v = np.vectorize(function_7)
outf = f7v(inx)
ax.plot(inx, outf, 'r')
ax.set_title('CP 6.1.1 plot')
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
fig.show

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
inx = np.arange(0, 1, 0.1)
f8v = np.vectorize(function_8)
outf = f8v(inx)
ax.plot(inx, outf, 'r')
ax.set_title('CP 6.1.1 plot 2nd function')
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
fig.show

```
